```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target, name="species")
```

```python
# Keep only 2 species: 1 = versicolor, 2 = virginica
X = X[y != 0]
y = y[y != 0]

# Regression ke liye ek feature target lenge (petal length ko predict karna hai)
y_reg = X['petal length (cm)']
X_reg = X.drop('petal length (cm)', axis=1)

# Polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_reg)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_poly, y_reg, test_size=0.2, random_state=42)

# Model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Polynomial Regression Results:")
print("Mean Squared Error (MSE):", round(mse, 4))
print("R² Score:", round(r2, 4))
```
